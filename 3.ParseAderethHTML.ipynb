```python
import pandas as pd
from bs4 import BeautifulSoup
import lxml

from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager

import requests
import numpy as np
import re
import time
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

# Visit visitcostarica.herokuapp.com
url = "http://adereth.github.io/oneoff/girltalk-v2/"
browser.visit(url)

time.sleep(1)

# Scrape page into Soup
html = browser.html
soup = BeautifulSoup(html, "html.parser")
```

```python
soup.find_all('div')
```

```python
artist_list = []
track_list = []

for div in soup.html.findAll(id=re.compile('track\d+')):
#     print(div['id'])
#     print(div)
    artist = div.find('h4').text
    track = div.find('p').text
    
    artist_list.append(artist)
    track_list.append(track)

print(artist_list)
print(track_list)
```

```python
len(artist_list) == len(track_list)
```

```python
track_list_stripped = [track.strip('"') for track in track_list]

songs_df = pd.DataFrame(list(zip(artist_list, track_list_stripped)), columns=['artist', 'track'])
```

```python
songs_df.to_csv('./data/artistAndTracks.csv')
```

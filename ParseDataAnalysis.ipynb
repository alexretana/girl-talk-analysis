```python
import pandas as pd
```

```python
allday_df = pd.read_json('./data/AllDay.json')

#drop first row because its the only empty track list
allday_df = allday_df[1:]
```

```python
trackList = []


for idx, row in allday_df.iterrows():
    if row.tracks != []:
        trackList.append(row.tracks[0]['link'])
        print(len(row.tracks[0]['link'])%2)
    
print(trackList)
```

```python
var = list(zip(list(allday_df[0::2].tracks.values), list(allday_df[1::2].tracks.values)))
```

```python
cnt = 0
data = []



for idx, song in allday_df.iterrows():
    song = song.tracks
    if not song:
        cnt += 1
        continue
    links = song[0]['link']
    for idxx, link in enumerate(links[0:-1:2]):
        linkType = link
        linkValue = links[idxx*2 + 1][1]
        data.append({
            'songID' : idx,
            'linkType': linkType,
            'linkValue' : linkValue
                })
            
parse_df = pd.DataFrame(data)
```

```python
parse_count_df = parse_df.groupby('songID').count()
parse_count_df
```
